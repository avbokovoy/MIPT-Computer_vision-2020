## Импорт необходимых модулей

```python
import cv2
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Чтение изображения из файла

```python
img = cv2.imread( '../../img/dragonfly.jpg', cv2.IMREAD_COLOR )
cv2.imshow("test",img)
plt.imshow( img )
```

## Запись изображения в файл

```python
cv2.imwrite( "../../img/my_dragonfly.jpg", img )
```

```python
print( type(img) )
print( img.shape )
print( img.dtype )
print( img[:2, :2, :] )
print( img.min() )
print( img.max() )
```
