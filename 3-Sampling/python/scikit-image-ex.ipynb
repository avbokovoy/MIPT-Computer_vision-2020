```python
from skimage import io
```

```python
dragonfly = io.imread( '../../img/dragonfly.jpg' )
io.imsave('./tmp.jpg', dragonfly, )
```

```python
plt.imshow(dragonfly)
```

```python
from skimage.util import img_as_ubyte, img_as_float
print(dragonfly)
```

```python
dragonfly = img_as_float(dragonfly)
print(dragonfly)
```

```python
dragonfly = img_as_ubyte( dragonfly )
print(dragonfly)
```

```python
gray_dragonfly = sk.color.rgb2gray( dragonfly )
plt.imshow(gray_dragonfly, cmap='gray')
print(gray_dragonfly)
```

```python
hsv_dragonfly = sk.color.rgb2hsv( dragonfly )
plt.imshow(hsv_dragonfly)
```

```python
astronaut = sk.data.astronaut()
plt.imshow(astronaut)
```

```python
astronaut = sk.util.invert(astronaut)
plt.imshow(astronaut)
```

```python
astronaut = sk.util.invert(astronaut)
hist = sk.exposure.histogram(astronaut[0])
```

```python
plt.hist(hist)
```

```python
contrasted = sk.exposure.rescale_intensity( astronaut )
plt.imshow(contrasted)
```

```python
gray_astronaut = sk.color.rgb2gray(astronaut)
equalized = sk.exposure.equalize_hist( gray_astronaut )
ada_equalized = sk.exposure.equalize_adapthist( gray_astronaut )
f, ax = plt.subplots(1,3)
ax[0].imshow(gray_astronaut, cmap='gray')
ax[1].imshow(equalized, cmap='gray')
ax[2].imshow(ada_equalized, cmap='gray') 
```

```python
from skimage import transform
resized = sk.transform.resize(gray_astronaut, (gray_astronaut.shape[0] // 4, gray_astronaut.shape[1] // 4), anti_aliasing = True)
plt.imshow(resized, cmap='gray')
```

```python
rescaled = sk.transform.rescale(gray_astronaut, 0.1, anti_aliasing=False)
plt.imshow(rescaled, cmap='gray')
```

```python
downscaled = sk.transform.downscale_local_mean( gray_astronaut, (4,3) )
plt.imshow(downscaled, cmap='gray')
```

```python
tform = sk.transform.AffineTransform(scale=(1.3, 1.1), rotation=1, shear=0.7, translation=(210, 50))
transformed_img = sk.transform.warp( gray_astronaut, tform )
plt.imshow(transformed_img, cmap='gray')
```
